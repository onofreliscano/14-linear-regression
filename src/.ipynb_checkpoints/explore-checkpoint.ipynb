```python
# BLOQUE 01
# SETUP DEL PROYECTO
# - Importar librerías base para EDA + ML
# - Configurar warnings para leer limpio
# - Fijar semilla para reproducibilidad
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

RANDOM_STATE = 42
```

```python
# BLOQUE 02
# CARGA DEL DATASET
# - Intentar cargar desde archivo local del repo
# - Si no existe, cargar desde el link oficial
# - Mostrar shape para validar que cargó bien
from pathlib import Path

local_path = Path("medical_insurance_cost.csv")
url = "https://breathecode.herokuapp.com/asset/internal-link?id=416&path=medical_insurance_cost.csv"

if local_path.exists():
    df = pd.read_csv(local_path)
else:
    df = pd.read_csv(url)

df.shape
# Resultado esperado:
# - Una tupla (filas, columnas) con 1000+ filas y 7 columnas
# Interpretación:
# - Dataset listo para EDA y modelado
```

```python
# BLOQUE 03
# VISTA RÁPIDA DE DATOS
# - Ver primeras filas para entender columnas
# - Confirmar presencia de 'charges' como target
# - Detectar valores raros visualmente
df.head()
# Resultado esperado:
# - Columnas: age, sex, bmi, children, smoker, region, charges
# Interpretación:
# - 'charges' será la variable objetivo (costo)
```

```python
# BLOQUE 04
# TIPOS DE DATOS Y NULOS
# - Revisar dtypes para saber qué es numérico/categórico
# - Contar nulos por columna
# - Preparar estrategia de limpieza
df.info()
df.isna().sum()
# Resultado esperado:
# - Pocos o cero nulos
# Interpretación:
# - Si hay nulos, toca imputar antes del modelo
```

```python
# BLOQUE 05
# DUPLICADOS Y VALORES ÚNICOS
# - Ver si hay filas duplicadas
# - Ver cardinalidad de categóricas (sex/smoker/region)
# - Prevenir ruido en entrenamiento
duplicates = df.duplicated().sum()

uniques = {
    "sex": df["sex"].nunique(),
    "smoker": df["smoker"].nunique(),
    "region": df["region"].nunique(),
}

duplicates, uniques
# Resultado esperado:
# - Duplicados: 0 o pocos
# - sex ~2, smoker ~2, region ~4
# Interpretación:
# - Categóricas manejables con OneHotEncoder
```

```python
# BLOQUE 06
# ESTADÍSTICAS DESCRIPTIVAS
# - Entender rangos y outliers rápidos
# - Ver dispersión de 'charges'
# - Identificar variables con escalas distintas
df.describe(include="all").T
# Resultado esperado:
# - Resumen numérico y categórico
# Interpretación:
# - 'charges' suele estar sesgada a la derecha (colas largas)
```

```python
# BLOQUE 07
# DISTRIBUCIÓN DEL TARGET (CHARGES)
# - Visualizar si hay sesgo fuerte
# - Decidir si conviene transformación log
# - Anticipar dificultad del modelo lineal
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.hist(df["charges"], bins=40)
plt.title("Distribución de charges")
plt.xlabel("charges")
plt.ylabel("frecuencia")
plt.show()

df["charges"].skew()
# Resultado esperado:
# - Histograma con cola a la derecha
# - skew positivo (normalmente alto)
# Interpretación:
# - Posible mejora usando log1p(charges) en optimización
```

```python
# BLOQUE 08
# RELACIÓN NUMÉRICAS VS TARGET
# - Ver relación rápida con scatter (linealidad)
# - Detectar patrones fuertes (age/bmi)
# - Confirmar si hay dispersión alta
num_cols = ["age", "bmi", "children"]

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for ax, col in zip(axes, num_cols):
    ax.scatter(df[col], df["charges"], s=10)
    ax.set_title(f"{col} vs charges")
    ax.set_xlabel(col)
    ax.set_ylabel("charges")
plt.tight_layout()
plt.show()

# Resultado esperado:
# - age y bmi con tendencia positiva, children más débil
# Interpretación:
# - Señales útiles, pero dispersión sugiere variables clave (smoker)
```

```python
# BLOQUE 09
# IMPACTO DE CATEGÓRICAS EN CHARGES
# - Comparar promedios por grupo (smoker suele dominar)
# - Medir diferencia de magnitudes por categorías
# - Priorizar features para el modelo
cat_cols = ["sex", "smoker", "region"]

group_means = {c: df.groupby(c)["charges"].mean().sort_values(ascending=False) for c in cat_cols}
group_means
# Resultado esperado:
# - smoker=yes con promedio mucho mayor
# Interpretación:
# - 'smoker' será feature crítica para pricing
```

```python
# BLOQUE 10
# CORRELACIÓN ENTRE NUMÉRICAS
# - Ver correlación para multicolinealidad básica
# - Confirmar si hay variables redundantes
# - Alinear con supuestos de regresión lineal
corr = df[["age", "bmi", "children", "charges"]].corr()
corr
# Resultado esperado:
# - Correlaciones moderadas, no perfectas
# Interpretación:
# - Baja probabilidad de multicolinealidad severa en numéricas
```

```python
# BLOQUE 11
# TRAIN/TEST SPLIT
# - Separar target y features
# - Split reproducible con random_state
# - Evitar fuga de datos (fit solo con train)
from sklearn.model_selection import train_test_split

X = df.drop(columns=["charges"])
y = df["charges"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE
)

X_train.shape, X_test.shape
# Resultado esperado:
# - 80/20 aprox
# Interpretación:
# - Base correcta para medir generalización del modelo
```

```python
# BLOQUE 12
# PIPELINE DE PREPROCESAMIENTO
# - OneHotEncoder para categóricas
# - StandardScaler para numéricas (mejora estabilidad)
# - ColumnTransformer para mantener todo limpio
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline

numeric_features = ["age", "bmi", "children"]
categorical_features = ["sex", "smoker", "region"]

preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore", drop=None), categorical_features),
    ]
)

# Resultado esperado:
# - Objeto 'preprocess' listo
# Interpretación:
# - Features quedan en formato apto para LinearRegression
```

```python
# BLOQUE 13
# MODELO BASE: LINEAR REGRESSION
# - Entrenar modelo por defecto
# - Pipeline: preprocess + model
# - Medir desempeño en test
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

baseline_model = Pipeline(
    steps=[
        ("preprocess", preprocess),
        ("model", LinearRegression()),
    ]
)

baseline_model.fit(X_train, y_train)
y_pred_base = baseline_model.predict(X_test)

mse_base = mean_squared_error(y_test, y_pred_base)
rmse_base = np.sqrt(mse_base)
r2_base = r2_score(y_test, y_pred_base)

mse_base, rmse_base, r2_base
# Resultado esperado:
# - r2 moderado/alto (depende del split)
# Interpretación:
# - Baseline define el punto de partida del pricing predictor
```

```python
# BLOQUE 14
# DIAGNÓSTICO RÁPIDO DE ERRORES (RESIDUALES)
# - Ver si el error crece con el valor (heteroscedasticidad)
# - Detectar patrones no lineales
# - Justificar optimización con log
residuals_base = y_test - y_pred_base

plt.figure(figsize=(6, 4))
plt.scatter(y_pred_base, residuals_base, s=10)
plt.axhline(0)
plt.title("Residuals vs Predicted (baseline)")
plt.xlabel("Predicted charges")
plt.ylabel("Residuals")
plt.show()

# Resultado esperado:
# - Dispersión creciente para predicciones altas (común)
# Interpretación:
# - Log-transform puede estabilizar varianza y mejorar ajuste
```

```python
# BLOQUE 15
# OPTIMIZACIÓN 1: TRANSFORMAR TARGET CON LOG1P
# - Mantener modelo lineal pero con target transformado
# - Reducir sesgo del target y estabilizar varianza
# - Comparar métricas vs baseline
from sklearn.compose import TransformedTargetRegressor

log_model = Pipeline(
    steps=[
        ("preprocess", preprocess),
        ("model", TransformedTargetRegressor(
            regressor=LinearRegression(),
            func=np.log1p,
            inverse_func=np.expm1
        )),
    ]
)

log_model.fit(X_train, y_train)
y_pred_log = log_model.predict(X_test)

mse_log = mean_squared_error(y_test, y_pred_log)
rmse_log = np.sqrt(mse_log)
r2_log = r2_score(y_test, y_pred_log)

mse_log, rmse_log, r2_log
# Resultado esperado:
# - RMSE baja o r2 sube (a menudo mejora)
# Interpretación:
# - Mejor predictor de costos cuando hay colas largas en charges
```

```python
# BLOQUE 16
# OPTIMIZACIÓN 2: RIDGE (REGULARIZACIÓN)
# - Seguir siendo regresión lineal, pero más robusta
# - Probar alfas con CV
# - Elegir el mejor por RMSE
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge_pipe = Pipeline(
    steps=[
        ("preprocess", preprocess),
        ("model", Ridge(random_state=RANDOM_STATE)),
    ]
)

param_grid = {"model__alpha": [0.1, 1.0, 10.0, 50.0, 100.0]}

grid = GridSearchCV(
    ridge_pipe,
    param_grid=param_grid,
    scoring="neg_root_mean_squared_error",
    cv=5,
    n_jobs=-1
)

grid.fit(X_train, y_train)

best_alpha = grid.best_params_["model__alpha"]
best_rmse_cv = -grid.best_score_

best_alpha, best_rmse_cv
# Resultado esperado:
# - alpha ganador y RMSE CV razonable
# Interpretación:
# - Ridge puede generalizar mejor si hay ruido/colinealidad leve
```

```python
# BLOQUE 17
# COMPARACIÓN FINAL DE MODELOS
# - Comparar baseline vs log-target vs ridge
# - Elegir el mejor por R2 y RMSE
# - Tomar decisión simple para negocio
results = pd.DataFrame([
    {"model": "baseline_linear", "rmse": rmse_base, "r2": r2_base},
    {"model": "log_target_linear", "rmse": rmse_log, "r2": r2_log},
    {"model": f"ridge_alpha_{best_alpha}", "rmse": best_rmse_cv, "r2": np.nan},
]).sort_values(by="rmse")

results
# Resultado esperado:
# - Tabla ordenada por rmse
# Interpretación:
# - Elegimos el modelo más estable para estimar premiums
```

```python
# BLOQUE 18
# ENTRENAR MODELO FINAL (DECISIÓN PRÁCTICA)
# - Si log_target gana en test, lo usamos
# - Si baseline es similar, preferimos baseline por simpleza
# - Dejar variable 'final_model' lista para guardar
final_model = log_model if rmse_log <= rmse_base else baseline_model
final_model.fit(X_train, y_train)

final_pred = final_model.predict(X_test)
final_rmse = np.sqrt(mean_squared_error(y_test, final_pred))
final_r2 = r2_score(y_test, final_pred)

type(final_model).__name__, final_rmse, final_r2
# Resultado esperado:
# - Nombre del modelo final + métricas finales
# Interpretación:
# - Modelo listo para usarse como estimador de primas
```

```python
# BLOQUE 19
# IMPORTANCIA DE FEATURES (COEFICIENTES) - APROX
# - Extraer nombres de features post-encoding
# - Mostrar top coeficientes absolutos
# - Entender drivers del premium (negocio)
ohe = final_model.named_steps["preprocess"].named_transformers_["cat"]
cat_feature_names = ohe.get_feature_names_out(categorical_features)

feature_names = numeric_features + list(cat_feature_names)

# Obtener coeficientes según el tipo de modelo
inner = final_model.named_steps["model"]
if isinstance(inner, TransformedTargetRegressor):
    coefs = inner.regressor_.coef_
    intercept = inner.regressor_.intercept_
else:
    coefs = inner.coef_
    intercept = inner.intercept_

coef_df = pd.DataFrame({
    "feature": feature_names,
    "coef": coefs
})
coef_df["abs_coef"] = coef_df["coef"].abs()
coef_df_sorted = coef_df.sort_values("abs_coef", ascending=False).head(10)

intercept, coef_df_sorted
# Resultado esperado:
# - Top 10 features por impacto
# Interpretación:
# - Verás fuerte señal de smoker y señales de edad/bmi
```

```python
# BLOQUE 20
# GUARDAR ARTEFACTO DEL MODELO
# - Guardar pipeline completo (preprocess + modelo)
# - Permite predicción reproducible en producción
# - Dejar en carpeta models/
import os
import joblib

os.makedirs("models", exist_ok=True)
model_path = "models/linear_regression_insurance.joblib"

joblib.dump(final_model, model_path)

model_path
# Resultado esperado:
# - Ruta del archivo guardado
# Interpretación:
# - Modelo listo para ser cargado y usado para cotizaciones
```

```python
# BLOQUE 21
# EXPORTAR PREDICCIONES DE EJEMPLO
# - Guardar tabla con y_real vs y_pred
# - Útil para QA del modelo y revisión rápida
# - Dejar en carpeta data/processed/
os.makedirs("data/processed", exist_ok=True)

pred_df = X_test.copy()
pred_df["charges_real"] = y_test.values
pred_df["charges_pred"] = final_pred
pred_df["error"] = pred_df["charges_real"] - pred_df["charges_pred"]

out_path = "data/processed/test_predictions.csv"
pred_df.to_csv(out_path, index=False)

out_path, pred_df.head()
# Resultado esperado:
# - CSV generado con predicciones
# Interpretación:
# - Evidencia tangible del rendimiento del modelo
```
